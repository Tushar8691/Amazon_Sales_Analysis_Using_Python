GROUP BY

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False) #here if we use ascending = True then it return small to large
```

```python
Size_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size',y='Qty',data=Size_Qty)
```

```python
Size_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

ax=sns.barplot(x='Size',y='Qty',data=Size_Qty)

for bar in ax.containers:
    ax.bar_label(bar)
```

Note :  From above graph we obserbe that most of the Qty buys M-Size in sales

```python
df['Category']=df['Category'].astype('str') # here we chage data type of category because it is in 'object'
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90) # here we use this for  labelling x-axis name in 90 degree
plt.show()
```

Note : From above graph we observe that most of the buyers are T-shirt

```python
#checking b2b data using  by pi chart  
B2B_check=df['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check,autopct='%1.1f%%')
plt.show()
```

```python
#checking b2b data using  by pi chart  
B2B_check=df['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
# plt.axis('equal')
plt.show()
```

Note : From above pie chart we observe that maximum i.e 99.2% are buyers are retailer and 0.8% are B2B  buyers

```python
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plot count of cites by states
plt.figure(figsize=(12,5))
sns.countplot(x='ship-state',data=df)
plt.xlabel('Ship-states')
plt.ylabel('Count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_state = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,5))
sns.countplot(x='ship-state',data=df[df['ship-state'].isin(top_10_state.index)])
plt.xlabel('Ship-states')
plt.ylabel('Count')
plt.title('Distribution of states')
plt.xticks(rotation=90)
plt.show()
```

Note : From above graph we observe that most buyers are from Maharashtra state

CONCLUSION

Based on the analysis of the data, we can conclude that most of the sales are made for M-Size t-shirts, with the majority of buyers being retail customers from Maharashtra state. A small fraction of buyers are B2B customers.
